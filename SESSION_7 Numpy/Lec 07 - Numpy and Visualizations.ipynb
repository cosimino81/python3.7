# Numpy and Visualizations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

#### First example

```python
# Creating the array
x = np.arange(2, 20)
y = 2 * x + 5
```

```python
# Set the figure size
plt.figure(figsize=(8,4))

# Plotting the data
plt.plot(x, y, 'b')

# Setting the title
plt.title('Matplot & Numpy Demo')

# Setting the axis names
plt.xlabel('Samples')
plt.ylabel('Values')
plt.show()
```

#### Little variation

```python
# Set the figure size
plt.figure(figsize=(8,4))

# Plotting the data
plt.plot(x, y, 'ob')

# Setting the title
plt.title('Matplot & Numpy Demo')

# Setting the axis names
plt.xlabel('Samples')
plt.ylabel('Values')
plt.show()
```

### Second Example: Exponential Plot

```python
# Calculate 2**x for all elements in the array.
x = np.arange(1,11)
y = np.exp2(x)
```

```python
# Make a new plot
plt.figure(figsize=(8,4))
plt.plot(x, y, 'og')
plt.title('Exponential Plot')
plt.xlabel('Samples')
plt.ylabel('Exponential')
plt.show()
```

```python
# Make a new plot
plt.figure(figsize=(8,4))
plt.bar(x, y)
plt.title('Exponential Plot')
plt.xlabel('Samples')
plt.ylabel('Exponential')
plt.ylim(-10, 1100)
plt.show()
```
