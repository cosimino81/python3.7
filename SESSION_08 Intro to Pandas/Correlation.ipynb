```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
df['SalePrice'].plot(kind='hist')
plt.show()
```

```python
import seaborn as sns
corrmat = df.corr()
corrmat.head()
```

```python
f, ax = plt.subplots(figsize =(10, 8)) 
sns.heatmap(corrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1)
```
